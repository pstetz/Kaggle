### Imports

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

### Load data

```python
mpi_region = pd.read_csv("../../data/kiva/kiva_mpi_region_locations.csv")
mpi_region.fillna(np.nan)
mpi_region.head(5)
```

### Drop columns

LocationName is just the combination of the country and region

geo is the combination of lat and lon

```python
mpi_region = mpi_region.drop(["LocationName", "geo"], axis=1)
```

```python
pd.DataFrame({
    "Total": len(mpi_region),
    "Missing": mpi_region.isnull().sum()
})
```

```python
fig, axarr = plt.subplots(3, figsize=(18, 18))

fig.subplots_adjust(hspace=0.8)

mpi_region.world_region.value_counts().head(10).plot(kind="bar", ax=axarr[0], title="World Region")

mpi_region.country.value_counts().head(10).plot(kind="bar", ax=axarr[1], title="Country")

_ = mpi_region.region.value_counts().head(10).plot(kind="bar", ax=axarr[2], title="Region")
```

```python
fig, axarr = plt.subplots(1, 2, figsize=(18, 6))

mpi_region.lat.plot(kind="hist", ax=axarr[0], title="Longitude")

_ = mpi_region.lat.plot(kind="hist", ax=axarr[1], title="Latitude")
```
