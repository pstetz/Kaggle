```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
%matplotlib inline
```

```python
train = pd.read_csv("../../data/amazon_employee/train.csv")
test  = pd.read_csv("../../data/amazon_employee/test.csv")
```

```python
train.describe()
```

```python
sns.set(rc={'figure.figsize':(14, 6)})

for col in train.columns:
    if col != "ACTION":
        pop_col = train[col].value_counts().head(10).keys()
        g = sns.barplot(x=col, y="ACTION", data=train[train[col].isin(pop_col)])
        g.set_ylabel("Action taken")
        plt.title("Action taken by {}".format(col))
        plt.show()
```
