```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

master_testing_data = pd.read_csv('../data/talking_data/test.csv')
```

```python
train = pd.read_csv('../data/talking_data/train.csv', nrows=1000000)
```

```python
downloaded_size = len(train[train['is_attributed']==1])
downloaded_size
```

```python
not_downloaded_size = len(train[train['is_attributed']==0])
not_downloaded_size
```

```python
percent_downloaded = downloaded_size / (downloaded_size + not_downloaded_size)
percent_downloaded * 100
```

```python
def bar_chart(feature):
    downloaded = train[train['is_attributed']==1][feature].value_counts()
    df = pd.DataFrame([downloaded])
    df.index = ['Yes']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
bar_chart('device')
```

```python
bar_chart('os')
```

```python
bar_chart('channel')
```

```python
bar_chart('app')
```

```python
#correlation matrix
corrmat = train.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8)
```
