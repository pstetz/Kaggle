```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Imputer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
```

```python
training_data = pd.read_csv('../data/talking_data/train.csv', nrows=100000)
```

```python
predictor_cols = ['app', 'device', 'os', 'channel']

y = training_data.is_attributed
x = training_data[predictor_cols]
ohe_x = pd.get_dummies(x)
train_x, val_x, train_y, val_y = train_test_split(ohe_x.as_matrix(), y.as_matrix(), test_size=0.25)
```

```python
pipeline = make_pipeline(Imputer(), LogisticRegression())

pipeline.fit(train_x, train_y)
```

```python
validation_results = pipeline.predict(val_x)

val_results_to_binary = m.to_discrete(list(validation_results), 0, 1)

# print("Mean Absolute Error : " + str(mean_absolute_error(validation_results, val_y)))
print("Percent correct : " + str(m.percent_right(val_results_to_binary, val_y)))
```

```python
predictions = pipeline.predict(test_x.as_matrix())
```
